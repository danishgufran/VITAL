```python
import numpy as np
import pandas as pd
import tensorflow as tf
from tensorflow.keras import layers
import tensorflow.keras.layers as L
import tensorflow_addons as tfa
import glob, random, os, warnings
import matplotlib.pyplot as plt
from sklearn.metrics import confusion_matrix, classification_report
import seaborn as sns
import cv2
total = []
temp = []
er = []

def flatten_list(_2d_list):
    flat_list = []
    # Iterate through the outer list
    for element in _2d_list:
        if type(element) is list:
            # If the element is of type list, iterate through the sublist
            for item in element:
                flat_list.append(item)
        else:
            flat_list.append(element)
    return flat_list


# ['BLU','HTC','LG','MOTO','OP3','S7']
devices = ['BLU','HTC','LG','MOTO','OP3','S7']
# ['engr0','engr1','glover','sciences','libstudy']
path = ['engr0']
for floorplan in path:
  print('*******************************************************************************************************')
  # print(floorplan)

  d = 'BLU'
  train_x = [cv2.imread(file) for file in glob.glob("content/train/"+str(floorplan)+ "_" + str(d) +"/*.png")]
  
  train_x = np.array(train_x)
  
  train_df, _, train_macs, lbl2cord = build_dataset(
    d,
    floorplan,
  )
  _, train_y = train_df[train_macs].values, train_df["label"].values

  print(train_x.shape, train_y.shape)

  for train_device in devices:

    
    temp_x = [cv2.imread(file) for file in glob.glob("content/train/"+str(floorplan)+ "_" + str(train_device) +"/*.png")]
    
    temp_x = np.array(temp_x)
    print(train_x.shape)
    print(temp_x.shape)
    # N = temp_x.shape[2] - train_x.shape[2]
    # temp_x = np.pad(temp_x, (0, N), 'constant')
    print(temp_x.shape)
    temp_x = temp_x/127
    
    train_df, _, train_macs, lbl2cord = build_dataset(
      train_device,
      floorplan,
    )
    _, temp_y = train_df[train_macs].values, train_df["label"].values

    train_x = np.concatenate((train_x, temp_x), axis=0)
    train_x = np.concatenate((train_x, temp_x*1.003), axis=0)
    train_x = np.concatenate((train_x, temp_x*2), axis=0)
    train_x = np.concatenate((train_x, temp_x*1.04), axis=0)
    train_x = np.concatenate((train_x, temp_x*3), axis=0)
    # train_x = np.concatenate((train_x, temp_x/1.9), axis=0)

    train_y = np.concatenate((train_y, temp_y), axis=0)
    train_y = np.concatenate((train_y, temp_y), axis=0)
    train_y = np.concatenate((train_y, temp_y), axis=0)
    train_y = np.concatenate((train_y, temp_y), axis=0)
    train_y = np.concatenate((train_y, temp_y), axis=0)
    # train_y = np.concatenate((train_y, temp_y), axis=0)
    

    train_x2 = [cv2.imread(file) for file in glob.glob("content/test/"+str(floorplan)+ "_" + str(train_device) + "_" + str(train_device) +"/*.png")]
    train_x2 = np.array(train_x2)
      # test_x = test_x/255
    _, test_df, test_macs, lbl2cord_test = build_dataset(
          train_device,
          floorplan,
      )
    missing_waps = list(set(train_macs) - set(test_macs))
    test_df = test_df.copy()  # supresses fragmented df warning
    test_df[missing_waps] = 0.0
    _ = test_df[train_macs].values.astype(float)
    train_y2 = test_df["label"].values.flatten()

    # train_x2 = [cv2.imread(file) for file in glob.glob("content/train/"+str(floorplan)+ "_" + str(train_device) +"/*.png")]
    # train_x2 = train_x2/127  
    train_x = np.concatenate((train_x, train_x2/10), axis=0)
    train_y = np.concatenate((train_y, train_y2), axis=0)
    
  #   train_x = np.concatenate((train_x, train_x), axis=0)
  #   train_x = np.concatenate((train_x, train_x2*1.003), axis=0)
  #   train_x = np.concatenate((train_x, train_x2*1.01), axis=0)
  #   train_x = np.concatenate((train_x, train_x*1.004), axis=0)
  #   train_x = np.concatenate((train_x, train_x*1.05), axis=0)
  #   train_x = np.concatenate((train_x, train_x* 1.02), axis=0)
  #   train_x = np.concatenate((train_x, train_x*1.04), axis=0)
  #   # train_x = np.concatenate((train_x, train_x), axis=0)

    
  #   train_df, _, train_macs, lbl2cord = build_dataset(
  #     train_device,
  #     floorplan,
  # )
  #   _, train_y = train_df[train_macs].values, train_df["label"].values

  #   train_y = np.concatenate((train_y, train_y), axis=0)
    
#   train_y = np.concatenate((train_y, train_y2), axis=0)
#   train_y = np.concatenate((train_y, train_y), axis=0)
#   train_y = np.concatenate((train_y, train_y), axis=0)
#   train_y = np.concatenate((train_y, train_y), axis=0)
#   train_y = np.concatenate((train_y, train_y), axis=0)
#   # train_y = np.concatenate((train_y, train_y), axis=0)
#   # train_y = np.concatenate((train_y, train_y), axis=0)

  train_x = np.resize(train_x,(train_y.shape[0],1,train_x.shape[2],3))
  print('---->',train_x.shape, train_y.shape)
  num_classes = int(max(train_y)) + 1
  input_shape = train_x.shape[1:]
  snd = 202
  div = 5
  learning_rate = 0.0001
  weight_decay = 0.0001
  batch_size = 256
  num_epochs = 500
  image_size =206 # We'll resize input images to this size
  patch_size = int(train_x.shape[-2]/div)  # Size of the patches to be extract from the input images
  num_patches = (image_size // patch_size) ** 2
  projection_dim = 25
  num_heads = 5 # 5 4 3 2 1
  transformer_units = [
      projection_dim * 2,
      projection_dim,
  ]  # Size of the transformer layers
  transformer_layers = 1
  mlp_head_units = [ 128, 64]  # Size of the dense layers of the final classifier
  ep = 0.000001

  data_augmentation = keras.Sequential(
  [
      layers.Normalization(),
      # layers.Dropout(0.1),
      layers.Resizing(image_size, image_size),
      # layers.GaussianNoise(0.1),
      
      # layers.RandomRotation(factor=1.0),
      # layers.RandomZoom(
      #     height_factor=1.0, width_factor=0.5
      # ),
  ],
  name="data_augmentation",
)
  # Compute the mean and the variance of the training data for normalization.
  data_augmentation.layers[0].adapt(train_x)

  def mlp(x, hidden_units, dropout_rate):
    for units in hidden_units:
        x = layers.Dense(units, activation=tf.nn.relu)(x)
        x = layers.Dropout(dropout_rate)(x)
    return x

  
  class Patches(layers.Layer):
    def __init__(self, patch_size):
        super(Patches, self).__init__()
        self.patch_size = patch_size

    def call(self, images):
        batch_size = tf.shape(images)[0]
        patches = tf.image.extract_patches(
            images=images,
            sizes=[1, self.patch_size, self.patch_size, 1],
            strides=[1, self.patch_size, self.patch_size, 1],
            rates=[1, 1, 1, 1],
            padding="VALID",
        )
        patch_dims = patches.shape[-1]
        patches = tf.reshape(patches, [batch_size, -1, patch_dims])
        return patches

    import matplotlib.pyplot as plt

  # plt.figure(figsize=(4, 4))
  image = train_x[np.random.choice(range(train_x.shape[0]))]
  # plt.imshow(image.astype("uint8"))
  # plt.axis("off")

  resized_image = tf.image.resize(
      tf.convert_to_tensor([image]), size=(image_size, image_size)
  )
  patches = Patches(patch_size)(resized_image)
  print(f"Image size: {image_size} X {202}")
  print(f"Patch size: {patch_size} X {patch_size}")
  print(f"Patches per image: {patches.shape[1]}")
  print(f"Elements per patch: {patches.shape[-1]}")

  n = int(np.sqrt(patches.shape[1]))
  plt.figure(figsize=(4, 4))
  for i, patch in enumerate(patches[0]):
      # ax = plt.subplot(n, n, i + 1)
      patch_img = tf.reshape(patch, (patch_size, patch_size, 3))
      # plt.imshow(patch_img.numpy().astype("uint8"))
      # plt.axis("off")


  class PatchEncoder(layers.Layer):
    def __init__(self, num_patches, projection_dim):
        super(PatchEncoder, self).__init__()
        self.num_patches = num_patches
        self.projection = layers.Dense(units=projection_dim)
        self.position_embedding = layers.Embedding(
            input_dim=num_patches, output_dim=projection_dim
        )

    def call(self, patch):
        positions = tf.range(start=0, limit=self.num_patches, delta=1)
        encoded = self.projection(patch) + self.position_embedding(positions)
        return encoded

  def create_vit_classifier():
    inputs = layers.Input(shape=input_shape)
    # Augment data.
    augmented = data_augmentation(inputs)
    # Create patches.
    patches = Patches(patch_size)(augmented)
    # Encode patches.
    encoded_patches = PatchEncoder(num_patches, projection_dim)(patches)

    # Create multiple layers of the Transformer block.
    for _ in range(transformer_layers):
        # Layer normalization 1.
        x1 = layers.LayerNormalization(epsilon=ep)(encoded_patches)
        # Create a multi-head attention layer.
        attention_output = layers.MultiHeadAttention(
            num_heads=num_heads, key_dim=projection_dim, dropout=0.01
        )(x1, x1)
        # Skip connection 1.
        x2 = layers.Add()([attention_output, encoded_patches])
        # Layer normalization 2.
        x3 = layers.LayerNormalization(epsilon=ep)(x2)
        # MLP.
        x3 = mlp(x3, hidden_units=transformer_units, dropout_rate=0.02)
        # Skip connection 2.
        encoded_patches = layers.Add()([x3, x2])

    # Create a [batch_size, projection_dim] tensor.
    representation = layers.LayerNormalization(epsilon=ep)(encoded_patches)
    representation = layers.Flatten()(representation)
    representation = layers.Dropout(0.04)(representation)
    # Add MLP.
    features = mlp(representation, hidden_units=mlp_head_units, dropout_rate=0.05)
    # Classify outputs.
    logits = layers.Dense(num_classes, activation = 'softmax')(features)
    # Create the Keras model.
    model = keras.Model(inputs=inputs, outputs=logits)
    return model

  def run_experiment(model):
    # optimizer = tfa.optimizers.AdamW(
    #     learning_rate=learning_rate, weight_decay=weight_decay
    # )
    optimizer = keras.optimizers.Adam(
        learning_rate=learning_rate
    )

    # optimizer = keras.optimizers.SGD(learning_rate=0.1)

    model.compile(
        optimizer=optimizer,
        loss=keras.losses.sparse_categorical_crossentropy,
        metrics=[
              keras.metrics.SparseCategoricalAccuracy(name="accuracy"),
            keras.metrics.SparseTopKCategoricalAccuracy(5, name="top-5-accuracy"),
        ],
    )

    checkpoint_filepath = "/tmp/checkpoint"
    checkpoint_callback = [keras.callbacks.ModelCheckpoint(
        checkpoint_filepath,
        monitor="loss",
        save_best_only=True,
        save_weights_only=True,
    ),
    tfa.callbacks.TQDMProgressBar(show_epoch_progress=False)]

    history = model.fit(
        x=train_x,
        y=train_y,
        batch_size=batch_size,
        epochs=num_epochs,
        # shuffle = True,
        callbacks=[checkpoint_callback],
        verbose = 1
    )

    # model.load_weights(checkpoint_filepath)
    # _, accuracy, top_5_accuracy = model.evaluate(test_x, test_y)
    # print(f"Test accuracy: {round(accuracy * 100, 2)}%")
    # print(f"Test top 5 accuracy: {round(top_5_accuracy * 100, 2)}%")

    return history



  vit_classifier = create_vit_classifier()
  history = run_experiment(vit_classifier)
  devices_test = ['BLU','HTC','LG','MOTO','OP3','S7']
  # path_test = ['engr0']
  # for floorplan_test in path_test:
  print('*******************************************************************************************************')
  # print(floorplan)
  for test_device in devices_test:
    test_x = [cv2.imread(file) for file in glob.glob("content/test/"+str(floorplan)+ "_" + str(train_device) + "_" + str(test_device) +"/*.png")]
    test_x = np.array(test_x)
    # test_x = test_x/255
    _, test_df, test_macs, lbl2cord_test = build_dataset(
        test_device,
        floorplan,
    )
    missing_waps = list(set(train_macs) - set(test_macs))
    test_df = test_df.copy()  # supresses fragmented df warning
    test_df[missing_waps] = 0.0
    _ = test_df[train_macs].values.astype(float)
    test_y = test_df["label"].values.flatten()
  
    pred = history.model.predict(test_x)
    # print(pred[0])
    pred_add = np.argmax(pred,axis =-1)
    # print(pred_add)
    
    md = ['i12p','pxl4','nk7']
    if test_device in md:

      train_df, test_df, macs, lbl2cords = build_dataset(
        md[0],
        floorplan
    )
    
    error = np.mean(
              compute_distances(
                  pred_add,
                  test_y
              )
          )
   # error = error/10
    print(str(floorplan)+ "_" + str(train_device) + "_" + str(test_device))
    print(pred_add)
    print(error)
    temp.append(error)
    # print(pred_add)
  er.append([temp])
  temp = []
err = flatten_list(er)

df = pd.DataFrame(err, columns= devices, index = path )
df.to_csv('dac_vit_no_dam.csv', encoding='utf-8', index = True)


fig, ax1 = plt.subplots(figsize=(5,5))
# fig, ax2 = plt.subplots(figsize=(15,5))
# fig, ax3 = plt.subplots(figsize=(25,5))
# fig, ax4 = plt.subplots(figsize=(35,5))
# fig, ax5 = plt.subplots(figsize=(45,5))
print(floorplan)

print(err)
print(np.mean(err))
x_axis_labels = ['i12p','pxl4','nk7','BLU','HTC','LG','MOTO','OP3','S7']# labels for x-axis
y_axis_labels = ['ENGR0', 'ENGR1', 'GLOVER', 'SCIENCES', 'LIBSTUDY'] # labels for y-axis
sb.heatmap(err, xticklabels=x_axis_labels, yticklabels=y_axis_labels,annot=True, square=True, ax=ax1, fmt='.4g')
# sb.heatmap(engr1, xticklabels=x_axis_labels, yticklabels=y_axis_labels,annot=True, square=True, ax=ax2, fmt='.4g',cmap="Blues", linecolor='black')
# sb.heatmap(glover, xticklabels=x_axis_labels, yticklabels=y_axis_labels,annot=True, square=True, ax=ax3, fmt='.4g',cmap="Blues", linecolor='black')
# sb.heatmap(sciences, xticklabels=x_axis_labels, yticklabels=y_axis_labels,annot=True, square=True, ax=ax4, fmt='.4g',cmap="Blues", linecolor='black')
# sb.heatmap(libstudy, xticklabels=x_axis_labels, yticklabels=y_axis_labels,annot=True, square=True, ax=ax5, fmt='.4g',cmap="Blues", linecolor='black')
plt.title("Train once ViT-LOC ")
plt.xlabel('Testing Devices',)
plt.ylabel('Floorplan',)
plt.yticks(rotation=0,fontsize=16);
plt.xticks(fontsize=12);
# plt.legend()
plt.tight_layout()
plt.savefig("Train once ViT-LOC")
err= []
temp = []
er = []
# err = []

```
